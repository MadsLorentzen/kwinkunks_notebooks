# Data Underground API

We just set up an instance of [CKAN](https://github.com/ckan/ckan). Let's have a look at how to use [the API](https://docs.ckan.org/en/2.8/api/).

## Natively

```python
import requests

url = "https://dataunderground.org/api/3/action/package_list"

re = requests.get(url)

re.json()['result']
```

```python
url = "https://dataunderground.org/api/3/action/package_search"

params = {'q': 'australia'}

re = requests.get(url, params)

re.json()
```

## Using `ckanapi`

The [`ckanapi`](https://github.com/ckan/ckanapi) command-line interface and Python package provides access to the API as well.

Install it with `pip`:

    pip install ckanapi

```python
from ckanapi import RemoteCKAN

uri = 'https://dataunderground.org'
ua = 'Matt Hall (+http://agilescientific.com/)'
ckan = RemoteCKAN(uri, user_agent=ua)

ckan.action.package_list()
```

```python
ckan.action.tag_list()
```

```python
q = ckan.action.package_search(q='production')
q
```

```python
csv = q['results'][0]['resources'][0]['url']
csv
```

```python
import pandas as pd

df = pd.read_csv(csv)
```
