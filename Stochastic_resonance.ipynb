```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(500)
y = 0.3 + np.sin(x/20.)/5
n = y + np.random.rand(500)/3 - (3./20)
```

```python
plt.figure(figsize=(10,3))

plt.subplot(121)
plt.plot(x, y, lw=2)
plt.text(10, 0.61, 'detection threshold')
plt.text(10, 0.05, "signal we'd like to detect", color='b')
plt.ylabel('amplitude')
plt.xlabel('time (ms)')
plt.ylim(0, 0.8)
plt.xlim(0, 500)
plt.hlines(0.6, 0, 500)

plt.subplot(122)
plt.plot(x, n, color='g', alpha=0.8)
plt.plot(x, y, color='b', lw=2)
plt.text(10, 0.69, "additive noise", color='g', alpha=0.8)
plt.xlabel('time (ms)')
plt.ylim(0,0.8)
plt.xlim(0, 500)
plt.hlines(0.6,0,500)

plt.savefig('stochastic_resonance_demo.png', dpi=625/10.)
plt.show()
```
