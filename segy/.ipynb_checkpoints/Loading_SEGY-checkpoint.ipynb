I have modified `segpy.py` a bit. It now reads data for ordinary floating point data, but the result looks bad (see below).

However, it fails completely for 'IBM floating point' numbers (e.g. in `ld0057_file_00095.sgy`), because `ibm2ieee2()` is not working for some reason. It complains that it needs a string of length 4, but as far as I can tell, that's what it should be getting. 

```python
#filename = 'HUN00-ALT-01_STK.sgy' # doesn't seem to work - data look bad
#filename = 'ld0057_file_00095.sgy' # loads vbut data don't look great
filename = '31_81_PR.sgy' # works!

with open(filename, 'rb') as f:
    data, sh, segyTraceHeaders = segypy.read_segy(f)
```

```python
vmin = np.percentile(data[data < 0], 1)
vmax = np.percentile(data[data > 0], 99)
```

```python
plt.figure(figsize=(18,24))
plt.imshow(data, aspect=0.5, vmin=vmin, vmax=vmax)
plt.show()
```
