# Sculpting horizons

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## 2D

Time is the second dimension of the section.

```python
arr = np.random.random((10, 15)) + 10

hor = np.random.randint(3, 7, size=10)

plt.imshow(arr.T)
plt.plot(hor, c='r')
```

```python
iline, time = arr.shape
idx = np.indices((time, iline))[0]
```

```python
result = np.moveaxis(arr, -1, 0) * (idx > hor)
result = np.moveaxis(result, 0, -1)
```

```python
plt.imshow(result.T, vmin=10, vmax=11)
plt.plot(hor, 'r')
plt.colorbar()
```

## 3D

Time on the last dimension.

```python
arr = np.random.random((10, 20, 30)) + 10

hor = np.random.randint(10, 20, size=(10, 20))

plt.imshow(arr[0].T)
plt.plot(hor[0], c='r')
plt.colorbar()
```

```python
iline, xline, time = arr.shape
idx = np.indices((time, iline, xline))[0]
```

```python
result = np.moveaxis(arr, -1, 0) * (idx > hor)
result = np.moveaxis(result, 0, -1)
```

```python
plt.imshow(result[0].T, vmin=10, vmax=11)
plt.plot(hor[0], c='r')
plt.colorbar()
```
